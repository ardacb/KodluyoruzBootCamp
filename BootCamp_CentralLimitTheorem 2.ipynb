## Merkezi Limit Teoremi

### Merkezi limit teoremi bize şunu söyler: 

### Bir popülasyonun dağılımına bakılmaksızın bu populasyondan rastgele alınan sample'ların (n>=30) ortalamasının populasyon ortalamasına yakınsayacağını ve normal dağılacağını ifade eder. Sample sayısı arttıkça da normal dağılıma daha da yaklaşır

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

rnd_list = np.random.randint(0,100,10000)
pop_mean=np.mean(rnd_list)
print('population mean=',pop_mean)
population=rnd_list

#0 ile 100 arasındaki sayılardan rastgele 10000 tane sayı generate edip ortalamasını alıyoruz
#populasyonu ve ortalamasını oluşturduk
#alıcağım samplelar'ın ortalamasının 49.2942 e yakın olması gerekiyor, bakalım öyle mi olacak
```

```python
#eleman sayısı 10,000 olan populasyonumuzdan 50'li samplelar almaya başlayalım
#sample sayımızı yavaş yavaş arttıralım
sample_means=[]
for i in range(1,10):
    sample_means.append(np.random.choice(rnd_list,size=50,replace=False).mean())

graph=sns.histplot(sample_means)
graph.set_xlabel('Sample Mean')
graph.set_title('n=10')
print('sample mean=',np.mean(sample_means))

```
